```python
import pandas as pd

```

```python
df = pd.read_csv("2Food_Delivery_Times.csv")
```

```python
df.dropna(subset=['Distance_km'], inplace=True)
df.dropna(subset=['Traffic_Level'], inplace=True)
df.dropna(subset=['Time_of_Day'], inplace=True)
df.dropna(subset=['Courier_Experience_yrs'], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
